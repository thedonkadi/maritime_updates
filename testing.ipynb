```python
import easyocr
import cv2
import matplotlib.pyplot as plt

# Initialize the EasyOCR reader
reader = easyocr.Reader(['en'])  # You can specify languages like ['en', 'fr', 'de']

# Load an image using OpenCV
image_path = 'handwritten_image_1.jpeg'
img = cv2.imread(image_path)

if img is None:
    print("Error: Could not load image.")
else:
    # Use EasyOCR to read text from the image
    result = reader.readtext(img)

    # Display results
    for (bbox, text, prob) in result:
        print(f"Detected text: {text} (Confidence: {prob:.2f})")
        
        # Draw bounding box around detected text
        (top_left, top_right, bottom_right, bottom_left) = bbox
        top_left = tuple(map(int, top_left))
        bottom_right = tuple(map(int, bottom_right))
        cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 2)

    # Show the image with detected text
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
import easyocr
import cv2

def extract_and_print_text(image_path):
    # Initialize the EasyOCR reader
    reader = easyocr.Reader(['en'])  # Specify languages as needed

    # Load the image
    img = cv2.imread(image_path)
    if img is None:
        print("Error: Could not load image.")
        return

    # Use EasyOCR to read text from the image
    result = reader.readtext(img)

    # Extract and concatenate all detected text
    full_text = " ".join([text for (_, text, _) in result])

    # Print the entire detected text
    print("Detected Text:\n", full_text)

# Example usage
image_path = 'uploads\Screenshot 1.png'
extract_and_print_text(image_path)
```

```python
# Define the path to the image
image_path = 'uploads\Screenshot 1.png' # Replace with your image file path

# Load image and prepare request
from paddleocr import PaddleOCR

# Initialize OCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Perform OCR
result = ocr.ocr(image_path, cls=True)
extracted_text = "\n".join([line[1][0] for line in result[0]])
print("Extracted Text:", extracted_text)

```
